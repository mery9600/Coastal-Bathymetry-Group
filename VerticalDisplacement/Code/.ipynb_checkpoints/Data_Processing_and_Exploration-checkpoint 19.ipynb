## GeoTiff processing

#### Below are steps for taking two geotiff scenes from the Alaska Satellite Facility API (covering the northern and southern halves of the outer banks), merging and exporting them with new metadata, and clipping using a shapefile converted from the original kml file.

```python
# import component files and reproject them to return rasters and transformations
from rasterio.warp import reproject

filepath = "../RawData/South_S1AA_20181008T230552_20181125T230551_VVP048_INT80_G_weF_61C9/S1AA_20181008T230552_20181125T230551_VVP048_INT80_G_weF_61C9_vert_disp.tif"
with rio.open(filepath) as south_vert_disp:
    south_vert = south_vert_disp.read(1)
    south_vert_meta = south_vert_disp.profile
    south_vert_reproj, south_vert_reproj_trans = reproject(source=rio.band(south_vert_disp, 1),
                                               dst_crs=south_vert_disp.profile['crs'],
                                               dst_resolution=(30, 30)
                                              )
filepath = "../RawData/North_S1AA_20181008T230617_20181125T230617_VVP048_INT80_G_weF_11AE/S1AA_20181008T230617_20181125T230617_VVP048_INT80_G_weF_11AE_vert_disp.tif"
with rio.open(filepath) as north_vert_disp:
    north_vert= north_vert_disp.read(1)
    north_vert_meta = north_vert_disp.profile
    north_vert_reproj, north_vert_reproj_trans = reproject(source=rio.band(north_vert_disp, 1),
                                               dst_crs=north_vert_disp.profile['crs'],
                                               dst_resolution=(30, 30)
                                              )
```

```python
from rasterio.io import MemoryFile
from rasterio.merge import merge
# create raster datasets from above transformed geotiffs

def create_dataset(data, crs, transform):
    # Receives a 2D array, a transform and a crs to create a rasterio dataset
    memfile = MemoryFile()
    dataset = memfile.open(driver='GTiff', height=data.shape[0], width=data.shape[1], count=1, crs=crs, 
                           transform=transform, dtype=data.dtype)
    dataset.write(data, 1)
        
    return dataset

south_vert_reproj_ds = create_dataset(south_vert_reproj[0], south_vert_disp.profile['crs'], south_vert_reproj_trans)
north_vert_reproj_ds = create_dataset(north_vert_reproj[0], north_vert_disp.profile['crs'], north_vert_reproj_trans)

merged, transf = merge([south_vert_reproj_ds, north_vert_reproj_ds])
# should i get rid of this next line? its the easiest way I could think of to reset the color scale so 0 is white
merged[0][0][0] = -1 * np.min(merged[0])
ep.plot_bands(merged[0],
              title="Complete Vertical Displacement in Outer Banks \n before and after Hurricane Michael (2018)",
              cmap="seismic")

plt.show()
```

```python
# create the metadata for the merged file, based on the format of the original component geotiffs
#print(north_vert_reproj_ds.meta)
#print(south_vert_reproj_ds.meta)

merged_meta = south_vert_reproj_ds.meta.copy()

# Update the metadata
merged_meta.update({"driver": "GTiff",
                 "height": merged.shape[1],
                 "width": merged.shape[2],
                 "transform": transf
                }
               )
print(merged_meta)
```

```python
out_file = "../RawData/10_08_to_11_25_combined.tif"
# for some reason this doesn't
with rio.open(out_file, "w", **merged_meta) as dest:
        dest.write(merged)
```
